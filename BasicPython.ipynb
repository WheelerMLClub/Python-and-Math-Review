# Python Review

##Classes

```
class Car():
  def __init__(self,wheels):
    self.wheels = wheelers
  def drive(self):
    if self.wheels <= 0:
      return False
    else:
      return True
```

Inheritance

```
class FormulaOneCar(Car):
  def __init__(self,wheels,color):
    super().__init__(wheels)
    self.color = color
  def drive(self):
    if self.wheels <= 4:
      return False
    else:
      return True
```
